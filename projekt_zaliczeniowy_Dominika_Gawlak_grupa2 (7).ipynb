```python
#Czƒô≈õƒá 3 - przetwarzanie jƒôzyka naturalnego
#PracujƒÖc na kolumnie text:
#1.Rozpoznaj osoby (persName) i przeka≈º je do kolumny o nazwie persons
```

```python
import spacy
```

```python
nlp=spacy.load("pl_core_news_lg")
```

```python
nlp.pipeline
```

```python
df['text'] = df['text'].apply(lambda text: nlp(text))
```

```python
print(df['text'])
type(df['text'][0])
```

```python
for doc in df['text']:
  for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
def extract_persons(text):
  doc = nlp(text)
  persons = [ent.text for ent in doc.ents if ent.label_ == 'persName']
  return ', '.join(persons)

df['persons'] = df['text'].apply(extract_persons)

print(df['persons'].head(50))
```

```python
#2.Rozpoznaj miejsca (placeName) i przeka≈º je do kolumny o nazwie places
```

```python
def extract_places(text):
  doc = nlp(text)
  places =[ ent.text for ent in doc.ents if ent.label_ == 'placeName']
  return ', '.join(places)

df['places'] = df['text'].apply(extract_places)

print(df['places'])
```

```python
#3.Rozpoznaj organizacje (orgName) i przeka≈º je do kolumny o nazwie organizations
```

```python
def extract_organizations(text):
  doc = nlp(text)
  organizations =[ ent.text for ent in doc.ents if ent.label_ == 'orgName']
  return ', '.join(organizations)

df['organizations'] = df['text'].apply(extract_organizations)

print(df['organizations'].head(50))
```
